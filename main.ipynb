```python
import gym
import matplotlib.pyplot as plt
import numpy as np
from agent import Agent
from train import Train
```

## Instantiating the environment and discovering the amount of states present and the number of possible actions

```python
env = gym.make('LunarLander-v2')
env.seed(0)

print('Number of actions: ', env.observation_space.shape[0])
print('Number of states: ', env.action_space.n)
```

```python
episodes=10
for episode in range  (1,episodes+1):
    state= env.reset()
    done=False
    score = 0

    while not done:
        env.render()
        action=env.action_space.sample()
        n_state,reward,done,info=env.step(action)
        score+=reward
    print('Episode :{} Score:{}'.format(episode,score))
env.close()
```

## Getting the agent, taking your training and saving the scores

```python
agent = Agent(state_size=8, action_size=4, seed=0, ddqn=False)
train = Train(env, agent)
scores, avg_scores = train.train()
```
